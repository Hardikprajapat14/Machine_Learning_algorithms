```python
# import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the dataset

df = pd.read_csv(r'C:\Mahesh\Datasets\Churn_Modelling.csv')

df.head()
```

```python
# information of the dataset

df.info()
```

```python
df.Exited.value_counts()
```

```python
df.Geography.unique()
```

```python
### Drop the irrelevant columns

df.drop(['RowNumber','CustomerId','Surname'],axis = 1,inplace=True)
```

```python
df.info()
```

```python
# one hot encoding

category_col = ['Geography','Gender']

ohe = pd.get_dummies(df[category_col],dtype=int,drop_first = True)

ohe
```

```python
df.head()
```

```python
df = pd.concat([df,ohe],axis=1)

df
```

```python
df.drop(category_col,axis=1,inplace=True)

df
```

```python
df.info()
```

```python
# independent and dependent features

x = df.drop('Exited',axis=1) # independent feature

y = df['Exited']
```

```python
x
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state= 42)

x_train
```

```python
y_train.value_counts()
```

```python
## Oversampling using SMOTE
from imblearn.over_sampling import SMOTE


# initialize
smt = SMOTE(sampling_strategy= 0.80,k_neighbors=7)

x_train_smt,y_train_smt = smt.fit_resample(x_train,y_train)
```

```python
y_train_smt.value_counts()
```

```python
### Ada Boost

from sklearn.ensemble import AdaBoostClassifier
```

```python
# initilize

abc = AdaBoostClassifier(n_estimators=60)
```

```python
abc_model = abc.fit(x_train_smt,y_train_smt)
```

```python
y_pred = abc_model.predict(x_test)
```

```python
## Evaluate the model

from sklearn.metrics import accuracy_score,classification_report
```

```python
print(f'Accuracy of ada boost classifier is : {accuracy_score(y_test,y_pred)}')
```
