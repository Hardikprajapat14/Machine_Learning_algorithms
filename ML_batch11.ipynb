```python
# dependent and independent features

x = df.drop('Exited',axis=1) # independent features
y = df.Exited # dependent feature
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
# Over sampling 

####1. random over sampler

from imblearn.over_sampling import RandomOverSampler
```

```python
# initialize

rs = RandomOverSampler(sampling_strategy=0.80)
```

```python
y_train.value_counts()
```

```python
x_train_rs,y_train_rs = rs.fit_resample(x_train,y_train)
```

```python
y_train_rs.value_counts()
```

```python
### SMOTE

from imblearn.over_sampling import SMOTE
```

```python
smt = SMOTE(sampling_strategy=0.80,k_neighbors=9)
```

```python
x_train_smt,y_train_smt = smt.fit_resample(x_train,y_train)
```

```python
y_train_smt.value_counts()
```
