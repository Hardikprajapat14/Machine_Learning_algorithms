```python
# dependent and independent features

x = df.drop('MEDV',axis=1) # independent features
y = df.MEDV # dependent feature
```

```python
x
```

```python
y
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
# model selection

from sklearn.ensemble import RandomForestRegressor
```

```python
# initialize

rfg = RandomForestRegressor()

# train
rfg.fit(x_train,y_train)

# y predict

y_pred = rfg.predict(x_test)
```

```python
# Accuracy of Random forest 

# Model evaluation

from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
print(f'Mean Square Error is : {mse}')
```

```python
y_test
```

```python
y_pred
```

```python
rmse = np.sqrt(mse)
print(f'Root Mean square error is : {rmse}')
```

```python
# R-Square 

r2 = r2_score(y_test,y_pred)
print(f'R-Square value is : {r2}')
```

```python
# Hyperparameter tuning 

from sklearn.model_selection import RandomizedSearchCV
```

```python
param = {
    'n_estimators' : [50,100,150,200,300],
    'criterion' : ['squared_error','absolute_error','friedman_mse', 'poisson'],
    'max_depth' : [5,10,12,15,None],
    'min_samples_split' : np.arange(2,10)
}
```

```python
rf_rscv = RandomizedSearchCV(rfg,param_distributions=param,cv=10,n_jobs=2,scoring='accuracy')
```

```python
rf_rscv.fit(x_train,y_train)
```

```python
rf_rscv.best_params_
```

```python
rfg_rscv = RandomForestRegressor(n_estimators=200,min_samples_split=6,max_depth=10,criterion='absolute_error')
```

```python
rfg_rscv.fit(x_train,y_train)
```

```python
y_pred_rf = rfg_rscv.predict(x_test)
```

```python
# Accuracy of Random forest 

# Model evaluation

from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(y_test,y_pred_rf)
print(f'Mean Square Error is : {mse}')
```

```python
rmse = np.sqrt(mse)
print(f'Root Mean square error is : {rmse}')
```

```python
# R-Square 

r2 = r2_score(y_test,y_pred_rf)
print(f'R-Square value is : {r2}')
```
